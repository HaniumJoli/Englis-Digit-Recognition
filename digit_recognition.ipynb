## Generate Dataset

```python
import cv2
import csv
import glob
header=["label"]
for i in range(0,4096):
    header.append("pixel"+str(i))
with open('digit_recognition.csv', 'a') as f:  
    writer = csv.writer(f)
    writer.writerow(header)
for level in range(10):
    image = "image/"

    dirList=glob.glob(image+str(level)+"/*.png")
    for img_path in dirList:
        im=cv2.imread(img_path)
        im_grey=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
        im_grey=cv2.GaussianBlur(im_grey,(15,15),0)
        roi=cv2.resize(im_grey,(64,64),interpolation=cv2.INTER_CUBIC)
        # cv2.imshow(roi)
        # cv2.waitKey(0)
        # cv2.destroyAllWindows()
        data=[]
        data.append(level)
        rows,cols=roi.shape
        # print(rows,cols)
        for i in range(rows):
            for j in range(cols):
                k=roi[i,j]
                if k<100:
                   k = 0
                else:
                    k = 1
                data.append(k)     
        with open('digit_recognition.csv', 'a') as f:    
              writer = csv.writer(f)
              writer.writerow(data)   
            
                    
    
```

```python
X=data.drop(["label"],axis=1)
Y=data["label"]
```

```python
##fitting model and setting the model
import joblib
from sklearn.svm import SVC
classifire=SVC(kernel='linear',random_state=6)
classifire.fit(X_train, Y_train)
joblib.dump(classifire, "model/digit_recognition")
```

```python
import tkinter as tk
from tkinter import messagebox
import os
import time
import pyscreenshot as ImageGrab
from PIL import Image
window = tk.Tk()
window.title("Handwritten Digit Recognition")
l1 = tk.Label(window, text="Digit Recognition", font=("Arial", 20))
l1.place(x=5, y=0)

t1 = tk.Entry(window, width=20, border=5)
t1.place(x=275, y=5)

def screen_capture():
    
    s1 = t1.get()
    
    if not s1:  
        messagebox.showerror("Input Error", "Please enter a folder name!")
        return
    

    os.startfile(r"C:/Users/USER/AppData/Local/Microsoft/WindowsApps/mspaint.exe")
    
    
    try:
        image_dir = r"D:/Projects/Python/digit_recognition/captured_images/" + s1
        if not os.path.exists(image_dir):
            os.makedirs(image_dir)
    except Exception as e:
        messagebox.showerror("Error", f"Failed to create directory: {str(e)}")
        return
    
    
    time.sleep(15) 
    
    
    image_folder = image_dir + "/"
    
    try:
        for i in range(5):  
            time.sleep(5)  
            im = ImageGrab.grab(bbox=(80, 250, 800, 800))  
            print("Saved......", i)
            im.save(image_folder + str(i) + '.png')

          
            img_preview = Image.open(image_folder + str(i) + '.png')
            img_preview.show()  
        
        print("Capture complete.")
        messagebox.showinfo("Result", "Capturing is completed!")
    except Exception as e:
        messagebox.showerror("Error", f"Failed during capture: {str(e)}")


b1 = tk.Button(window, text="1. Open and Paint", font=("Algerian", 15), bg="Blue", fg="black", command=screen_capture)
b1.place(x=5, y=40)
def generate_dataset():
    import cv2
    import csv
    import glob
    header=["label"]
    for i in range(0,4096):
        header.append("pixel"+str(i))
    with open('digit_recognition.csv', 'a') as f:  
        writer = csv.writer(f)
        writer.writerow(header)
    for level in range(10):
        image = "image/"
    
        dirList=glob.glob(image+str(level)+"/*.png")
        for img_path in dirList:
            im=cv2.imread(img_path)
            im_grey=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
            im_grey=cv2.GaussianBlur(im_grey,(15,15),0)
            roi=cv2.resize(im_grey,(64,64),interpolation=cv2.INTER_CUBIC)
            # cv2.imshow(roi)
            # cv2.waitKey(0)
            # cv2.destroyAllWindows()
            data=[]
            data.append(level)
            rows,cols=roi.shape
            # print(rows,cols)
            for i in range(rows):
                for j in range(cols):
                    k=roi[i,j]
                    if k<100:
                       k = 0
                    else:
                        k = 1
                    data.append(k)     
            with open('digit_recognition.csv', 'a') as f:    
                  writer = csv.writer(f)
                  writer.writerow(data)   
        messagebox.showinfo("Result", "Dataset is Generated!")      
                
                        
    
    

b2 = tk.Button(window, text="2. Generate Dataset", font=("Algerian", 15), bg="Sky Blue", fg="black",command=generate_dataset)
b2.place(x=5, y=90)
def train_save_accuracy():
    import pandas as pd
    from sklearn.utils import shuffle
    from sklearn.model_selection import train_test_split
    from sklearn.svm import SVC
    from sklearn import metrics
    import joblib
    from tkinter import messagebox

    # Load and shuffle the dataset
    try:
        data = pd.read_csv('digit_recognition.csv', low_memory=False)
    except Exception as e:
        messagebox.showerror("Error", f"Failed to load dataset: {str(e)}")
        return

    # Ensure the data types are correct
    try:
        data = shuffle(data)
        
        # Check for non-numeric values in the label column
        # Filter out any rows where the label is not a number (if it somehow contains non-numeric data)
        data = data[pd.to_numeric(data["label"], errors='coerce').notnull()]

        # Convert feature columns to float and label column to int
        X = data.drop("label", axis=1).astype(float)  # Convert all feature columns to float
        Y = data["label"].astype(int)  # Ensure label column is integer
    except Exception as e:
        messagebox.showerror("Error", f"Data processing error: {str(e)}")
        return

    # Split the data into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    # Train the SVM model
    classifier = SVC(kernel='linear', random_state=6)
    classifier.fit(X_train, Y_train)

    # Save the model
    model_dir = "model/"
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)
    joblib.dump(classifier, os.path.join(model_dir, "digit_recognition_model.pkl"))

    # Calculate and display accuracy
    prediction = classifier.predict(X_test)
    acc = metrics.accuracy_score(Y_test, prediction)
    messagebox.showinfo("Result", f"Accuracy is: {acc:.2f}")

    return acc


# No need to return anything since the accuracy is shown in the message box



    

b3 = tk.Button(window, text="3. Fix and model, calculate accuracy", font=("Algerian", 15), bg="purple", fg="black",command=train_save_accuracy)
b3.place(x=5, y=140)
def prediction():
    # Prediction of image drawn in Paint
    import joblib
    import cv2
    import numpy as np
    import time
    import pyscreenshot as ImageGrab
    import os

    # Start Paint
    os.startfile(r"C:/Users/USER/AppData/Local/Microsoft/WindowsApps/mspaint.exe")
    
    # Load the model
    try:
        model = joblib.load('model/digit_recognition_model.pkl')  # Ensure correct path and filename
    except Exception as e:
        print(f"Error loading model: {e}")
        return

    images_folder = "img/"
    if not os.path.exists(images_folder):
        os.makedirs(images_folder)

    time.sleep(10)  # Adjust this if needed
    while True:
        time.sleep(5)
        print("Capturing image: ")
        img = ImageGrab.grab(bbox=(80, 250, 800, 800))
        img.save(images_folder + "img.png")
        print("Saved image.")
        
        print("Reading image:")
        im = cv2.imread(images_folder + "img.png")
        im_grey = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        im_grey = cv2.GaussianBlur(im_grey, (15, 15), 0)
        
        # Threshold the image
        ret, im_th = cv2.threshold(im_grey, 100, 255, cv2.THRESH_BINARY)
        roi = cv2.resize(im_th, (64, 64), interpolation=cv2.INTER_AREA)
        
        rows, cols = roi.shape
        X = []
        
        # Converting image to features
        for i in range(rows):
            for j in range(cols):
                k = roi[i, j]
                X.append(0 if k < 100 else 1)

        predictions = model.predict([X])
        print("Prediction:", predictions[0])
        
        cv2.putText(im, "Prediction is: " + str(predictions[0]), (20, 20), 0, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
        cv2.startWindowThread()
        cv2.namedWindow("Result")
        cv2.imshow("Result", im)
        cv2.waitKey(0)
        
        cv2.destroyAllWindows()
        break


b4 = tk.Button(window, text="4. Live prediction", font=("Algerian", 15), bg="Violet", fg="black",command=prediction)
b4.place(x=5, y=190)


window.geometry("600x300")
window.mainloop()
```
